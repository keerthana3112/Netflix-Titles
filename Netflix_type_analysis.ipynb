```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("netflix.xlsx", sheet_name = "netflix_titlesoriginal")
df
```

```python
data = df[['show_id','type','duration']]
data
```

```python
new = data['duration'].str.split(" ", expand = True)
data.loc[:,'Duration'] = new[0]
data.loc[:,'Unit'] = new[1]
data.drop(columns = ['duration','Time'], inplace = True)
data
```

```python
movies_data = data[data['type'] == 'Movie']
movies_data
```

```python
%matplotlib inline
```

```python
x = movies_data['Duration'].apply(pd.to_numeric).values
fig1 , ax1 = plt.subplots(figsize = (10,20))
ax1.set_title('Movie Duration')
ax1.boxplot(x , notch = True)
```

```python
y = shows_data['Duration'].apply(pd.to_numeric).values
fig2 , ax2 = plt.subplots(figsize = (10,20))
ax2.set_title('TV shows Duration')
ax2.hist(y)
```
